## 2. Frequent category imputation

```python
# creating a copy of the original dataset
df = originalDF.copy()
```

```python
# checking how many data are missing in the column 'mort_acc'
_mv.hasMissingValues('emp_title', df)
```

> comment : The 'emp_title' column has 5.79% missing values.

```python
# counting the category that the column has
df['emp_title'].value_counts()
```

```python
# finding the frequent category
categories = df['emp_title'].value_counts()
frequentCategory = categories.keys()[0]
```

```python
frequentCategory
```

```python
# filling the missing values with frequent category
df['emp_title'].fillna(value=frequentCategory, inplace=True)
```

```python
# checking the column has any missing values left
_mv.hasMissingValues('emp_title', df)
```

comment : The column 'emp_title' has no missing values. Missing values are filled with frequent category.

```python
# counting the category after imputation
df['emp_title'].value_counts()
```

comment : The 'emp_title' column now has same number of category as before. But the frequent category now has more count of 22927 which is equal to the number of missing values before imputation.
