## 2. Drop Columns

```python
# creating a copy of the original dataset
df = originalDF.copy()
```

```python
# checking the count of missing values
_mv.countMissingValues(df, (20,8))
```

> comment : The dataframe has six columns with missing data.

<br>

The `'mort_acc'` column will be used in this illustration to handle missing data.

```python
# checking how many data are missing in the column 'mort_acc'
_mv.hasMissingValues('mort_acc', df)
```

> comment : The 'mort_acc' column has 9.54% missing values. Which is sufficient, and so the entire column can be deleted.

```python
# deleting the entire 'mort_acc' column from the dataset
df.drop('mort_acc', axis=1, inplace=True)
```

```python
# checking the count of missing values
_mv.countMissingValues(df, (20,8))
```

> comment : The dataframe now has five columns with missing data. The 'mort_acc' column is deleted.
